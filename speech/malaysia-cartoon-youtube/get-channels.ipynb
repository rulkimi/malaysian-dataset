```python
from bs4 import BeautifulSoup
from tqdm import tqdm
from glob import glob
import time
import os

files = glob('kartun melayu - YouTube.html')
```

```python
channels = set()

for f in tqdm(files):
    with open(f) as fopen:
        t = fopen.read()
        
    soup = BeautifulSoup(t, 'html.parser')

    links = soup.find_all("a", class_="yt-simple-endpoint style-scope yt-formatted-string")

    for link in links:
        channels.add(link['href'])
```

```python
len(channels)
rejected = [
    'TylerOliveira',
    'business',
    'entreprenuership',
    'optimumtech',
    'IT',
    'EthanWong',
    'TODAY',
    'lexfridman',
    'lebih',
    'Business',
    'hardknocks',
    'Christy',
    'https://www.youtube.com/@1983ncl',
    'PETRO',
    'https://www.youtube.com/@TVLine',
    'heavyspoilers',
    'XueDao'
]
filtered_channels = [c for c in channels if '@' in c if all([r not in c for r in rejected])]
len(filtered_channels)
```

```python
import time
```

```python
for x_ in tqdm(filtered_channels):
    name = x_.split('@')[1]
    f = f'{name}.txt'
    
    if os.path.exists(f) and os.stat(f).st_size > 30:
        continue
    
    command = f"yt-dlp --verbose --flat-playlist --print id '{x_}' > {f}"
    os.system(command)
    
    time.sleep(120)
```
